```python
one = df.loc[df["Class"] == 1];
zero = df.loc[df["Class"] == 0];
```

```python
counts,bin_edges=np.histogram(one['Var'],bins=10,density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
onecdf=plt.plot(bin_edges[1:],cdf,label="CDF Class 1")
counts,bin_edges=np.histogram(zero['Var'],bins=10,density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
twocdf=plt.plot(bin_edges[1:],cdf,label="CDF Class 0")
plt.xlabel("Var")
plt.ylabel("Probability")
plt.title("CDF of Nodes of Classes 1 and 0")
plt.legend()
plt.show()
```
