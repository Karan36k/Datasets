```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Mall_Customers.csv')
```

```python
df=df.drop([u'CustomerID'],axis=1)
```

```python
df[u'Gender'].value_counts()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue=u'Spending Score (1-100)', height=6) \
.map(plt.scatter, u'Annual Income (k$)', u'Spending Score (1-100)') \
.add_legend();
plt.show();
```

```python
sns.set_style("whitegrid")
sns.pairplot(df, hue=u'Spending Score (1-100)', height=4);
plt.show()
```
